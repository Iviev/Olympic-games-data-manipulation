### 10 pt: Which countries have more female medalists than male medalists?

````python
medal_list = athlete_events[athlete_events['Medal'].notnull()]
medal_list
female_group = medal_list[medal_list.Sex == 'F'].groupby('NOC').count()
female_group
female_medals = female_group['Medal']
female_medals 

male_group = medal_list[medal_list.Sex == 'M'].groupby('NOC').count()
male_group
male_medals = male_group['Medal']
male_medals


medal_df = pd.DataFrame(data=dict(medalist_f = female_medals, medalist_m = male_medals), index = female_medals.index)
medal_df

medal_df['difference_count'] = medal_df['medalist_f'] - medal_df['medalist_m']
medal_df['difference_count'].sort_values(ascending=False).head(10)
````

CHN has the highest number of female medalist. ROU, JAM, BLR, and UKR have a higher number of female medalist compared to other countries. 

# 10 pt: Which country won the most medal points in each of the 2014 Winter Olympics?
Rank each country by a score, where 3 points are assigned for a gold medal, 2 points for silver, and 1 point for bronze.

````python
event_2014 = athlete_events[athlete_events["Season"] == "Winter"]
event_2014 = athlete_events[(athlete_events.Season == 'Winter') & (athlete_events.Year == 2014) & ((athlete_events.Medal=="Gold") | (athlete_events.Medal=="Silver") | (athlete_events.Medal=="Bronze"))]
event_2014.groupby(['NOC'])['Medal'].size()
#group y noc and medal,empty dict make dict out of data frame, for item in my dictionary that was my 


medals_dict = event_2014.to_dict('index')
noc_dict = {}

for item in medals_dict:
    if medals_dict[item]['NOC'] not in noc_dict:
        noc = medals_dict[item]['NOC']
        if medals_dict[item]['Medal'] == 'Bronze':
            noc_dict[noc] = 1
        if medals_dict[item]['Medal'] == 'Silver':
            noc_dict[noc] = 2
        if medals_dict[item]['Medal'] == 'Gold':
            noc_dict[noc] = 3
    else:
        noc = medals_dict[item]['NOC']
        if medals_dict[item]['Medal'] == 'Bronze':
            noc_dict[noc] += 1
        if medals_dict[item]['Medal'] == 'Silver':
            noc_dict[noc] += 2
        if medals_dict[item]['Medal'] == 'Gold':
            noc_dict[noc] += 3
# country_dict

noc_medals = pd.DataFrame.from_dict(noc_dict, orient='index')
noc_medals.columns = ['Medal Points']
noc_medals.sort_values('Medal Points', ascending=False)








````

Canada won the most medal points in the 2014 Winter OLYMPICS
